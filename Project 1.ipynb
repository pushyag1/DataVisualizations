Scatter Plots: The coordinates of each point are defined by two dataframe columns and filled circles are used to represent each point. This kind of plot is useful to see complex correlations between two variables. Points could be for instance natural 2D coordinates like longitude and latitude in a map or, in general, any pair of metrics that can be plotted against each other.

```python
#plotting a scatter plot against Budget invested in the movie and the Domestic gross earned by a particular film.
fig, ax = plt.subplots()
ax.scatter(data['Budget($M)'], data['Domestic Gross($M)'])
ax.set_title('Movies Dataset for budget and domestic gross(without plot values)')
ax.set_xlabel('Budget($M)')
ax.set_ylabel('Domestic Gross($M)')



#citation:https://stackoverflow.com/questions/32319166/highlight-last-data-point-in-scatter-plot-with-pandas
#citation:https://towardsdatascience.com/the-next-level-of-data-visualization-in-python-dd6e99039d5e
#citation:https://plot.ly/python/line-and-scatter/
```

positive correlation : Variables that are positively correlated move in the same direction

```python
#plotting a scatter plot against Budget invested in the movie and the Domestic gross earned by a particular film.
#scatter plots with the values
plt.clf()

# using Budget and Domestic gross
xs = (data['Budget($M)'])
ys = (data['Domestic Gross($M)'])
plt.figure(figsize=(10,10))
plt.scatter(xs,ys,s=70, c='darkred')

# zip joins x and y coordinates in pairs
for x,y in zip(xs,ys):

    label = "{:.0f}".format(y)

    plt.annotate(label,# this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.xticks(np.arange(0,500,100))
plt.yticks(np.arange(0,1000,100))
plt.title('Movies Dataset(A scatter plot for budget and the domestic gross)(scatterplot with plot values)')
plt.xlabel('Budget of the movie in millions')
plt.ylabel('Domestic Gross of the movie in millions')

plt.show()

#Citation: https://pythonprogramming.net/legends-titles-labels-matplotlib-tutorial/
#Citation: https://stackoverflow.com/questions/29188757/matplotlib-specify-format-of-floats-for-tick-lables
#Citation: http://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
#citation:https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html
```

The above scatter plot hows us the Budget invested in the movie and the Domestic gross earned. It is observed that few movies despite of investing less made a large amount of profits, whereas few who invested more havent, on the plot we can observe that the movie which invested for about 300 million dollars has earned a profit of 937 million dollars.
The above scatterplot gives the complete description of the invested Budget and the earned profits.
The above scatterplot gives the complete description of the invested Budget and the earned domestic profits.

A histogram is a representation of the distribution of data. This function calls matplotlib.pyplot.hist() , on each series in the DataFrame, resulting in one histogram per column.

Histogram for the Worldwide Gross and domestic gross  in millions 

```python
#Histogram for the Worldwide Gross in millions
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(data['Worldwide Gross($M)'])
# set title and labels
ax.set_title('Movies Dataset')
ax.set_ylabel('Movies Count')
ax.set_xlabel('Worldwide Gross($M)')
data.hist(column="Worldwide Gross($M)", bins=5)
plt.title('Movies Dataset(A scatter plot for Movie count and the Domestic profits earned)')
plt.ylabel('Number of movies')
plt.xlabel('Domestic Gross of the movie in millions')

#Citation:https://wellsr.com/python/python-create-histogram-from-pandas-dataframe/
#Citation:https://www.mathworks.com/matlabcentral/answers/436999-histogram-x-axis-range
#Citation:https://stackoverflow.com/questions/6986986/bin-size-in-matplotlib-histogram
#Citation:https://towardsdatascience.com/the-next-level-of-data-visualization-in-python-dd6e99039d5e
```

The above histogram shows the profit earned (Worldwide gross) for the number of movies in the data set, The maximum profit earned by the the number of movies(movies count). more than 5000 movies earned a profit ranging from (0 to 500 million dollars) and fewer than 500 movies has earned a profit ranging from (600 to 1200 million dollars), fewer than 100 movies has earned a profit of about 1500 million dollars
