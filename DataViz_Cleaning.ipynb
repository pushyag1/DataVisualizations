```
df= pd.read_excel("https://www2.census.gov/programs-surveys/demo/tables/geographic-mobility/2005/state-to-state-migration/state_to_state_migrations_table_2005.xls",header=[5], index_col = 0)
df.head(10)
```

```
df = df.iloc[:-9]
df.tail(10)
```

```
df = df.dropna(axis = 0, how = 'all')
df.head()
```

```
df=df.drop(columns=df.columns[(df == 'MOE').any()])
df.head()
```

```
df1= df[df.columns[3:-5]]
df1.head()
```

```
new_header = df1.iloc[0] #grab the first row for the header
df1 = df1[1:] #take the data less the header row
df1.columns = new_header
df1.head()
```
